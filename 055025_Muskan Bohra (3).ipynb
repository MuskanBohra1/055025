### Categorical Data

1. **Count**: The total number of occurrences of each category in the dataset, providing a basic understanding of distribution among categories.

2. **Frequency**: The number of times each category appears in the dataset, allowing for easy comparison between categories.

3. **Proportion**: The fraction of the total that each category represents, calculated by dividing the count of each category by the total count, useful for understanding relative sizes.

4. **Minimum**: The lowest value in a categorical dataset, often relevant in ordered categories, indicating the least significant category.

5. **Maximum**: The highest value in a categorical dataset, highlighting the most significant category when categories are ordered.

6. **Mode**: The category that appears most frequently in the dataset, providing insight into the most common observation.

7. **Rank**: The order of categories based on a specific criterion, allowing for comparison of categories relative to one another.

8. **Correlation**: Measures the strength and direction of the relationship between two categorical variables, often assessed through contingency tables.

### Chi-squared Test

9. **Goodness of Fit**: A statistical test used to determine if observed frequencies match expected frequencies for a single categorical variable, assessing how well the data fits a specified distribution.

10. **Test of Independence**: A statistical test that evaluates whether two categorical variables are independent or associated, often using a contingency table to assess relationships between variables.

### Non-Parametric Tests

11. **Mann-Whitney**: A non-parametric test that compares differences between two independent groups based on ranks, suitable for ordinal data or non-normally distributed continuous data.

12. **Wilcoxon**: A non-parametric test that compares two related samples or matched groups, assessing differences in ranks without assuming normality.

13. **Kruskal-Wallis**: A non-parametric test that compares three or more independent groups based on ranks, useful when assumptions of ANOVA are not met.

14. **Friedman**: A non-parametric test that compares three or more related groups, assessing differences in ranks across multiple observations.

### Regression with Categorical Variables

15. **Logistic Regression**: A statistical method used to model binary outcomes based on one or more predictor variables, allowing for prediction of the probability of a particular outcome occurring.

### Visualizations

16. **Bar Chart**: A graphical representation of categorical data using bars to show the frequency or proportion of each category, making it easy to compare different groups visually.

17. **Histogram**: A graphical representation of the distribution of numerical data, where bars represent frequency counts for ranges of values, providing insight into data distribution.

18. **Pie Chart**: A circular chart divided into slices, illustrating the proportion of each category relative to the whole, helpful for visualizing categorical data distribution.

```python
# Performing a t-test to compare the means of import and export values in a dataset using Python.

from scipy import stats

# Subsetting data for imports and exports
import_data = data[data['Import_Export'] == 'Import']['Value']
export_data = data[data['Import_Export'] == 'Export']['Value']

# t-test to compare means of import vs export values
t_stat, p_value = stats.ttest_ind(import_data, export_data)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
# Chi-squared test of between Independence of Country and Import_Export.

country_import_export_crosstab = pd.crosstab(data['Country'], data['Import_Export'])
chi2_stat, p_val, dof, expected = stats.chi2_contingency(country_import_export_crosstab)

print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_val)
```

```python
# Use the Mann-Whitney U Test to compare the distributions of import and export values in a dataset.

from scipy.stats import mannwhitneyu

# Separate the two groups
import_values = data[data['Import_Export'] == 'Import']['Value']
export_values = data[data['Import_Export'] == 'Export']['Value']

# Mann-Whitney U Test
stat, p_value = mannwhitneyu(import_values, export_values)

print(f'Mann-Whitney U Test statistic: {stat}, p-value: {p_value}')
```

```python
# Use non-parametric tests like Mann-Whitney U, Wilcoxon, and Kruskal-Wallis to compare groups based on different variables in a dataset

from scipy.stats import mannwhitneyu, wilcoxon, kruskal, friedmanchisquare

# 1. Mann-Whitney U Test
# Compare Quantity between Air and Sea shipping methods
shipping_air = data[data['Shipping_Method'] == 'Air']['Quantity']
shipping_sea = data[data['Shipping_Method'] == 'Sea']['Quantity']  # Assuming 'Sea' exists

mannwhitneyu_result = mannwhitneyu(shipping_air, shipping_sea)
print('Mann-Whitney U Test between Air and Sea Shipping:', mannwhitneyu_result)

# 2. Wilcoxon Signed-Rank Test
# Compare Quantity for Air Shipping before and after April 2023
before_date_air = data[(data['Date'] < '2023-04-01') & (data['Shipping_Method'] == 'Air')]['Quantity']
after_date_air = data[(data['Date'] >= '2023-04-01') & (data['Shipping_Method'] == 'Air')]['Quantity']

# Ensure both samples are the same size (Wilcoxon requires paired data)
min_len_air = min(len(before_date_air), len(after_date_air))
wilcoxon_result = wilcoxon(before_date_air.iloc[:min_len_air], after_date_air.iloc[:min_len_air])
print('Wilcoxon Signed-Rank Test for Air Shipping before and after April 2023:', wilcoxon_result)

# 3. Kruskal-Wallis H Test
# Compare Quantity across different Product categories (e.g., Machinery, Electronics, Clothing)
product_categories = ['Machinery', 'Electronics', 'Clothing']
product_groups = [data[data['Category'] == product]['Quantity'] for product in product_categories]

kruskal_result = kruskal(*product_groups)
print('Kruskal-Wallis H Test across Product Categories:', kruskal_result)
```

```python
# Check the size of each group first

date_group1_air = data[(data['Date'] == '07-12-2023') & (data['Shipping_Method'] == 'Air')]['Quantity']
date_group2_air = data[(data['Date'] == '04-04-2023') & (data['Shipping_Method'] == 'Air')]['Quantity']
date_group3_air = data[(data['Date'] == '21-08-2024') & (data['Shipping_Method'] == 'Air')]['Quantity']

# Check the number of entries in each group

print("Group 1 size:", len(date_group1_air))
print("Group 2 size:", len(date_group2_air))
print("Group 3 size:", len(date_group3_air))

# Ensure that all groups have values, and the test can proceed

if len(date_group1_air) > 1 and len(date_group2_air) > 1 and len(date_group3_air) > 1:
    friedman_result = friedmanchisquare(date_group1_air, date_group2_air, date_group3_air)
    print('Friedman Test for Air Shipping across different dates:', friedman_result)
else:
    print("Error: One or more groups do not have sufficient data for the Friedman test.")
```

### Non-Categorical Data

**Measures of Central Tendency**  
- **Minimum**: The smallest value in the dataset, indicating the lowest point of observation.  
- **Maximum**: The largest value in the dataset, representing the highest point of observation.  
- **Mean**: The arithmetic average calculated by summing all values and dividing by the number of observations.  
- **Median**: The middle value in a sorted dataset, providing a measure that is less affected by outliers than the mean.  
- **Mode**: The value that appears most frequently in the dataset, useful for identifying the most common observation.  
- **Percentile**: A measure indicating the value below which a given percentage of data falls, helping to understand distribution.

**Measures of Dispersion**  
- **Range**: The difference between the maximum and minimum values, reflecting the spread of the data.  
- **Standard Deviation**: A statistic that measures how much individual data points deviate from the mean, indicating data variability.  
- **Skewness**: A measure of the asymmetry of the data distribution, indicating whether data points are skewed left or right.  
- **Kurtosis**: A statistic that measures the "tailedness" of the data distribution, indicating the presence of outliers.  
- **Correlation**: A measure that describes the strength and direction of the relationship between two variables.

**Composite Measure**  
- **Coefficient of Variation**: A standardized measure of dispersion calculated as the ratio of the standard deviation to the mean, expressed as a percentage, useful for comparing variability across datasets.  
- **Confidence Interval**: A range of values that likely contains the true population parameter, providing a measure of estimation uncertainty.

**Test of Mean**  
- **t-test**: A statistical test used to compare the means of two groups to determine if they are significantly different from each other.  
- **ANOVA**: Analysis of variance, used to compare the means of three or more groups to assess if at least one differs significantly.

**Test of Variance**  
- **F-test**: A statistical test used to compare the variances of two groups to determine if they are significantly different.  
- **Levene's Test**: Assesses whether two or more groups have equal variances, useful for validating ANOVA assumptions.  
- **Bartlett's Test**: A test for equal variances across multiple groups, sensitive to departures from normality.

**Test of Proportion**  
- **z-test**: A statistical test used to determine if the proportion in a single sample significantly differs from a known population proportion.  
- **Chi-squared test**: A statistical test that assesses the association between categorical variables by comparing observed and expected frequencies.

**Test of Correlation**  
- **t-test for correlation**: Used to assess the significance of the correlation coefficient, determining if the observed correlation between two variables is statistically significant.

**Test of Normality**  
- **Shapiro-Wilk**: A statistical test used to determine if a dataset is normally distributed, based on sample size and data characteristics.  
- **Kolmogorov-Smirnov**: Compares the sample distribution to a reference distribution to assess goodness-of-fit and normality.  
- **Anderson-Darling**: A test for normality that gives more weight to the tails of the distribution, making it sensitive to deviations.  
- **Jarque-Bera**: Tests for normality by evaluating skewness and kurtosis, assessing whether a dataset follows a normal distribution.

**Non-Parametric Tests**  
- **Mann-Whitney**: A non-parametric test that compares the differences between two independent groups without assuming normality.  
- **Wilcoxon**: A non-parametric test that compares two related samples, assessing whether their population mean ranks differ.  
- **Kruskal-Wallis**: A non-parametric test used to compare three or more independent groups when normality cannot be assumed.  
- **Friedman**: A non-parametric test for comparing three or more related groups, assessing differences in ranks across multiple observations.

**Regression**  
- **Cross-sectional Data**: Observations collected at a single point in time, useful for analyzing relationships and effects at that moment.  
- **Time-series Data**: Observations collected over time at regular intervals, ideal for analyzing trends and patterns across different time periods.  
- **Panel Data**: A combination of cross-sectional and time-series data, allowing for more comprehensive analyses by observing multiple subjects over time.

**Polynomial Regression**  
- **Cross-sectional Data**: A regression model that fits a polynomial equation to cross-sectional data, capturing non-linear relationships between variables.

**Scatter Plot**: A graphical representation of two continuous variables, displaying points to reveal potential relationships and trends.  
**Line Plot**: A visualization that connects data points with lines, effectively illustrating trends or changes over time.  
**Box-Whisker Plot**: A graphic that summarizes data distribution, showing the median, quartiles, and potential outliers, useful for comparing distributions.

```python
# Test of Variance {F | Levene | Bartlett}
import pandas as pd
from scipy import stats

# Test functions for variance comparison
def variance_tests(data, group_column, value_column, test_name):
    groups = [group[value_column].values for _, group in data.groupby(group_column)]
    
    if test_name == 'Levene':
        stat, p_value = stats.levene(*groups)
    elif test_name == 'Bartlett':
        stat, p_value = stats.bartlett(*groups)
    return stat, p_value

# F-Test for 'Quantity' between 'Import' and 'Export'
import_export_f_stat, import_export_p_value = stats.levene(
    data[data['Import_Export'] == 'Import']['Quantity'],
    data[data['Import_Export'] == 'Export']['Quantity']
)

# Levene's Test for 'Weight' across 'Product' categories
levene_stat, levene_p_value = variance_tests(data, 'Product', 'Weight', 'Levene')

# Bartlett's Test for 'Value' across 'Product' categories
bartlett_stat, bartlett_p_value = variance_tests(data, 'Product', 'Value', 'Bartlett')

# Structured output
output = f"""
F-Test (Quantity: Import vs Export):
  F-Statistic: {import_export_f_stat:.4f}, p-Value: {import_export_p_value:.4e}

Levene's Test (Weight by Product):
  Levene-Statistic: {levene_stat:.4f}, p-Value: {levene_p_value:.4e}

Bartlett's Test (Value by Product):
  Bartlett-Statistic: {bartlett_stat:.4f}, p-Value: {bartlett_p_value:.4e}
"""

# Print results
print(output)
```

```python
#  The normality of the 'Quantity' column using Shapiro-Wilk, Kolmogorov-Smirnov, Anderson-Darling, and Jarque-Bera tests.

from scipy import stats

# Select the 'Quantity' column for testing
quantity = data['Quantity']

# 1. Shapiro-Wilk Test
shapiro_test = stats.shapiro(quantity)

# 2. Kolmogorov-Smirnov Test
ks_test = stats.kstest(quantity, 'norm')

# 3. Anderson-Darling Test
anderson_test = stats.anderson(quantity)

# 4. Jarque-Bera Test
jarque_bera_test = stats.jarque_bera(quantity)

# Display the results in a structured format
print("Statistical Test Results:")
print("-" * 30)

print(f"1. Shapiro-Wilk Test:")
print(f"   Statistic: {shapiro_test.statistic:.4f}")
print(f"   p-value: {shapiro_test.pvalue:.4f}")

print("-" * 30)

print(f"2. Kolmogorov-Smirnov Test:")
print(f"   Statistic: {ks_test.statistic:.4f}")
print(f"   p-value: {ks_test.pvalue:.4f}")

print("-" * 30)

print(f"3. Anderson-Darling Test:")
print(f"   Statistic: {anderson_test.statistic:.4f}")
print(f"   Critical Values: {anderson_test.critical_values}")
print(f"   Significance Level: {anderson_test.significance_level}")

print("-" * 30)

print(f"4. Jarque-Bera Test:")
print(f"   Statistic: {jarque_bera_test.statistic:.4f}")
print(f"   p-value: {jarque_bera_test.pvalue:.4f}")
```
